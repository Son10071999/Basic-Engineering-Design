## Measuring skewness and kurtosis


**[TODO]** First, make a numeric_columns by selecting only columns containing
numerical data

```python
# numericColumns = df.describe().columns
# numericColumns
# Select columns with numerical data (excluding NaN columns)
numeric_columns = df.select_dtypes(include=['number'])

# Display the first few rows of the numeric_columns DataFrame
print(numeric_columns.head())
```

**[TODO]** Measure skewness for all numeric attribute and sort by value.

```python
# Calculate skewness for each numeric attribute
skewness = numeric_columns.apply(lambda x: x.skew())

# Sort the skewness values in desc order
sorted_skewness = skewness.sort_values(ascending=False)

# Display the sorted skewness values
print(sorted_skewness)
```

**[TODO]** Measure kurtosis for all numeric attribute and sort by value.

```python
# Calculate kurtosis for each numeric attribute
kurtosis_values = numeric_columns.apply(lambda x: x.kurtosis())

# Sort the kurtosis values in descending order
sorted_kurtosis = kurtosis_values.sort_values(ascending=False)

# Display the sorted kurtosis values
print(sorted_kurtosis)
```

# Statistical Hypothesis Test
**[TODO]** We would like the test that is there a significant price difference between `small` and `midsize` type car.

Test our hypothesis with proper statistical test with siginificance level as 0.05.

If there is missing value (`nan`) in `pd.Series`, test will not be working properly.

```python
from scipy.stats import ttest_ind

# Set the significance level
significantLevel = 0.05

# Filter the data for small and midsize cars, and remove rows with missing 'Price' values
smallCar = df[df['Type'] == 'Small']['Price'].dropna()
midsizeCar = df[df['Type'] == 'Midsize']['Price'].dropna()

# Perform a two-sample t-test only if both groups have valid price data
if not smallCar.empty and not midsizeCar.empty:
    t_statistic, p_value = ttest_ind(smallCar, midsizeCar)

    # Print the p-value
    print('P-value:', format(p_value, ".19f"))

    # Check if the p-value is less than the significance level
    if p_value < significantLevel:
        print("There is a significant price difference between small and midsize cars.")
    else:
        print("There is no significant price difference between small and midsize cars.")
else:
    print("Cannot perform the t-test due to missing price values in one of the groups.")
```

[TODO] We would like the test that is there a significant price difference between `USA Car` and `non-USA Car`.

Test our hypothesis with proper statistical test with siginificance level as `0.05`.

Please pay attention to the results of the hypothesis.




```python
# Set the significance level
significantLevel = 0.05

# Filter the data for USA cars and non-USA cars, and remove rows with missing 'Price' values
USA_Car_Price = df[df['Origin'] == 'USA']['Price'].dropna()
non_USA_Car_Price = df[df['Origin'] != 'USA']['Price'].dropna()

# Perform a two-sample t-test
t_statistic, p_value = ttest_ind(USA_Car_Price, non_USA_Car_Price)

# Print the p-value
print('P-value:', format(p_value, ".19f"))

# Check if the p-value is less than the significance level
if p_value < significantLevel:
    print("There is a significant price difference between USA cars and non-USA cars.")
else:
    print("There is no significant price difference between USA cars and non-USA cars.")
```
